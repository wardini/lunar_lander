```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
# Moon 2
N=40
dx1=30*np.random.random(N)
dx2=30*np.random.random(N)
x=np.zeros(2*dx1.shape[0]+2)
x[N]=-25
x[N+1]=25
for i in range(N):
    x[N-1-i]=x[N-i]-dx1[i]
    x[N+2+i]=x[N+1+i]+dx2[i]
y=40*np.random.random(x.shape[0])-10
y[N]=-0.01
y[N+1]=-0.01
```

```python
plt.figure()
plt.plot(x,y)
plt.gca().set_aspect(1.0) 
plt.show()    
```

```python
x = np.append(x,[10000,10000,-10000,-10000])
y = np.append(y,[   0,-10000,-10000,0])
print(list(zip(x,y)))
```

```python
# Asteroid
N=100
R=1000
dR=140
Lz=np.array([[1000,-25],[1000,25]])
adelta=np.random.random(N)
rdelta=dR*np.random.random(N)-dR/2
adelta= adelta*(2*np.pi-np.arctan(50/1000)-np.arctan(10/1000))/np.sum(adelta)
last_angle=np.arctan(25/1000)
x = np.zeros(N+2)
y = np.zeros(N+2)
x[0]=Lz[0,0]
x[1]=Lz[1,0]
y[0]=Lz[0,1]
y[1]=Lz[1,1]
for i in range(N):
    last_angle += adelta[i]
    x[i+2]=(R+rdelta[i])*np.cos(last_angle)
    y[i+2]=(R+rdelta[i])*np.sin(last_angle)
print(list(zip(x,y)))
```

```python
plt.figure()
plt.plot(x,y)
plt.gca().set_aspect(1.0) 
plt.show()    
```

```python
# Double Asteroid
# Asteroid #1 which has the landing zone
N=100
R=1000
dR=140
Lz=np.array([[1000,-25],[1000,25]])
adelta=np.random.random(N)
rdelta=dR*np.random.random(N)-dR/2
adelta= adelta*(2*np.pi-np.arctan(50/1000)-np.arctan(10/1000))/np.sum(adelta)
last_angle=np.arctan(25/1000)
x = np.zeros(N+2)
y = np.zeros(N+2)
x[0]=Lz[0,0]
x[1]=Lz[1,0]
y[0]=Lz[0,1]
y[1]=Lz[1,1]
for i in range(N):
    last_angle += adelta[i]
    x[i+2]=(R+rdelta[i])*np.cos(last_angle)
    y[i+2]=(R+rdelta[i])*np.sin(last_angle)
x=x-2000

# Asteroid #2
N=100
R=1000
dR=140
adelta=np.random.random(N)
rdelta=dR*np.random.random(N)-dR/2
adelta= adelta*2*np.pi/np.sum(adelta)
last_angle=np.arctan(25/1000)
x2 = np.zeros(N)
y2 = np.zeros(N)
for i in range(N):
    last_angle += adelta[i]
    x2[i]=(R+rdelta[i])*np.cos(last_angle)
    y2[i]=(R+rdelta[i])*np.sin(last_angle)
x2=x2+2000

print('[')    
print(list(zip(x,y)))
print(',')
print(list(zip(x2,y2)))
print(']')
```

```python
plt.figure()
plt.plot(x,y)
plt.plot(x2,y2)
plt.gca().set_aspect(1.0) 
plt.show()    
```
